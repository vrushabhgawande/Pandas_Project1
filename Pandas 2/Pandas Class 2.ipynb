```python
a = pd.read_csv('data2.csv')
a.dropna(inplace = True)
print(a.to_string())
```

```python
# replacing empty value
a = pd.read_csv('data2.csv')
a.fillna(42,inplace = True)
print(a)
```

```python
# replacing empty value for DS_Marks
a = pd.read_csv('data2.csv')
a["DS_Marks"].fillna(49,inplace = True)
a["Python_Marks"].fillna(50,inplace=True)
a["DA_Marks"].fillna(49, inplace=True)
print(a)
```

```python
# using mean
a = pd.read_csv('data2.csv')
b = a["DS_Marks"].mean()
a["DS_Marks"].fillna(b,inplace = True)
print(a)
```

```python
# Calulate medin
a = pd.read_csv('data2.csv')
b = a["Python_Marks"].median()
a["Python_Marks"].fillna(b,inplace = True)
print(a)
```

```python
#calculte mode
a = pd.read_csv('data2.csv')
b = a["Python_Marks"].mode()[0]
a["Python_Marks"].fillna(b,inplace = True)
print(a)
# note mode = the value that appears most frequently
```

```python
# changing data in specific row
a = pd.read_csv("data2.csv")
a.loc[1,'Python_Marks'] = 99
print(a.to_string())
```

```python
# replacce value greater than 100 by 95
a = pd.read_csv('data2.csv')
for x in a.index:
    if a.loc[x,"Python_Marks"] > 100:
        a.loc[x,"Python_Marks"] = 95
print(a.to_string())        
```

```python
a = pd.read_csv('data2.csv')
for x in a.index:
    if a.loc[x,"DS_Marks"] > 100:
        a.drop(x,inplace = True)
print(a.to_string())  

# Note do include the inplace = True argument to maake the changes in the oribginal 
# DataFrame object instedad of returning a copy
```
